```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cookie_cats.csv")
```

```python
# Oyuncuları round sayısına göre segment
df['player_segment'] = pd.cut(df['sum_gamerounds'],
                              bins=[0, 20, 100, df['sum_gamerounds'].max()],
                              labels=['casual', 'mid', 'core'])
```

```python
df['player_segment']
```

```python
seg_ret1 = df.groupby('player_segment', observed=True )['retention_1'].mean()
seg_ret7 = df.groupby('player_segment' , observed=True )['retention_7'].mean()
```

```python
print("Segmentlere göre 1. gün retention:\n", seg_ret1)
print("Segmentlere göre 7. gün retention:\n", seg_ret7)
```

### Göreselleştirme 

```python
plt.figure(figsize=(6,4))
sns.barplot(x=seg_ret1.index, y=seg_ret1.values)
plt.title("Segmentlere Göre 1. Gün Retention")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x=seg_ret7.index, y=seg_ret7.values)
plt.title("Segmentlere Göre 7. Gün Retention")
plt.show()
```
